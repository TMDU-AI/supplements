# matplotlib

matplotlibは、Pythonの主要なデータ可視化ライブラリの一つで、2次元図表と図を生成するために広く使用されています。このライブラリはNumPyを基盤として動作し、科学的な計算ツールキットと密接に連携します。matplotlibはその汎用性と機能の豊富さから、科学技術計算やデータ分析での使用に特に適しており、Pythonのデータサイエンス分野での標準的なツールの一つです。

まずは、matplotlibを使ってシンプルな折れ線グラフを作成する方法から説明します。

**matplotlibのインポートと基本的な折れ線グラフの描画**

```python
# matplotlib.pyplotをpltという名前でインポートします
import matplotlib.pyplot as plt

# グラフにプロットするデータを準備します
x = [1, 2, 3, 4, 5]  # x軸のデータ
y = [1, 4, 9, 16, 25]  # y軸のデータ（xの値を二乗したもの）

# plt.plot()を使って線グラフを作成します
plt.plot(x, y)

# グラフを表示します
plt.show()
```

このコードは、x と y のデータを使って基本的な線グラフを描きます。plt.show() は作成したグラフを画面に表示するために必要です。

**グラフのタイトルとラベルの追加**

次に、グラフにタイトルと軸ラベルを追加してみましょう。

```python
# グラフにタイトルと軸ラベルを追加
plt.plot(x, y)
plt.title("Simple Plot")  # グラフのタイトル
plt.xlabel("X axis")  # X軸のラベル
plt.ylabel("Y axis")  # Y軸のラベル

# グラフを表示
plt.show()
```

これで、グラフにタイトルとX軸、Y軸のラベルが追加されました。

**グラフのスタイルと色のカスタマイズ**

グラフの見た目をカスタマイズするために、線のスタイルや色を変更します。

```python
# 線のスタイルと色を変更してグラフを作成
plt.plot(x, y, color='red', linestyle='--')  # 赤色の点線
plt.title("Styled Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")

# グラフを表示
plt.show()
```

ここでは、color と linestyle オプションを使って、線の色とスタイルを変更しています。

**複数のデータセットのプロット**

複数のデータセットを同じグラフに表示することで、比較を容易にします。

```python
# 二つのデータセットを準備
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 3, 6, 11, 20]

# 複数の線をプロット
plt.plot(x, y1, label='x squared')  # xの二乗
plt.plot(x, y2, label='x times two plus one')  # xに2をかけて1を足したもの

# 凡例を追加
plt.legend()

# タイトルと軸ラベル
plt.title("Multiple Data Sets")
plt.xlabel("X axis")
plt.ylabel("Y axis")

# グラフを表示
plt.show()
```

**グリッドの追加**

グリッドを追加することで、データポイントの位置を把握しやすくなります。

```python
# グリッドを表示
plt.plot(x, y1, label='x squared')
plt.plot(x, y2, label='x times two plus one')
plt.title("Plot with Grid")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)  # グリッドの表示を有効にする
plt.legend()

# グラフを表示
plt.show()
```

**グラフの保存**

作成したグラフを画像ファイルとして保存することができます。

```python
# グラフを保存
plt.plot(x, y1, label='x squared')
plt.plot(x, y2, label='x times two plus one')
plt.title("Saved Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid(True)

# グラフをPNGファイルとして保存
plt.savefig('plot.png')

# 保存した後にグラフを表示
plt.show()
```

**ここまで使用した関数の説明と引数の種類**



*   plt.plot()   
この関数は、データの線グラフや点グラフを作成します。

   *   x, y: プロットするデータ。x は横軸の値、y は縦軸の値です。
   *   color: 線の色を指定します。

  *   linestyle: 線のスタイルを指定します（例：'-', '--', '-.', ':'）。

  *   marker: データ点にマーカーを設置する場合のスタイル（例：'o', 's', '^'）。

  *   label: 凡例に表示するラベル名。
*   plt.title()  
グラフにタイトルを追加します。

   *   label: タイトルのテキスト。
   *   fontdict: フォントに関する属性を辞書型で指定します。

  *   loc: タイトルの位置（'left', 'right', 'center'）。

*   plt.xlabel(), plt.ylabel()  
それぞれx軸とy軸にラベルを追加します。

   *   xlabel, ylabel: 軸のラベルテキスト。
   *   fontdict: フォント設定を辞書型で指定。

  *   labelpad: ラベルと軸との間のパディング（距離）。

   
*   plt.legend()  
凡例を追加します。

   *   loc: 凡例の位置を指定（'upper left', 'upper right', 'lower left', 'lower right' など）。
   *   shadow: 凡例に影をつけるかどうかの真偽値。

  *   fontsize: 凡例のフォントサイズ。
*   plt.grid()  
グリッド（格子）を表示します。

   *   b: グリッドを表示するかどうかの真偽値（True/False）。
   *   which: グリッド線を'major', 'minor', 'both' のどれに適用するか。

  *   axis: 'x', 'y', 'both' どの軸にグリッドを表示するか。
  *   color, linestyle, linewidth: グリッド線の色、スタイル、太さ。
*   plt.show()  
グラフを表示します。この関数は引数を取りません。
*   plt.savefig()
グラフを画像ファイルとして保存します。

   *   fname: 保存するファイル名。
   *   dpi: 解像度（dots per inch）。

  *   format: 画像形式（'png', 'jpg', 'svg', 'pdf' など）。
  *   bbox_inches: 保存する画像の範囲を制限。


**散布図 (plt.scatter())**

散布図は、2つの連続変数の関係を点で表現するグラフです。

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='blue')  # 点の色を青に設定
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

このコードは、x と y のデータ点を青色で散布図にプロットします。

**棒グラフ (plt.bar())**

カテゴリカルなデータの大小を棒の長さで比較するグラフです。

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

plt.bar(categories, values, color='green')  # 棒の色を緑に設定
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

このコードでは、異なるカテゴリーの値を緑色の棒グラフで表示しています。

**ヒストグラム (plt.hist())**

データの分布を視覚化します。特にデータの頻度（度数）分布を示すのに適しています。

```python
import numpy as np

data = np.random.normal(0, 1, 1000)  # 平均0、標準偏差1の正規分布から1000個のデータを生成

plt.hist(data, bins=30, color='purple')  # 30のビン（棒）で表示、色は紫
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

このヒストグラムは、生成されたデータの分布を30のビンで表示しています。

**円グラフ (plt.pie())**

データの全体に対する比率を扇形の面積で表現するグラフです。

```python
sizes = [215, 130, 245, 210]
labels = ['Apple', 'Banana', 'Cherry', 'Date']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # 円グラフが円になるようにアスペクト比を調整
plt.title("Pie Chart")
plt.show()
```

このコードでは、各要素のサイズ（数値）に応じて円グラフの扇形が割り当てられ、ラベルとパーセンテージが表示されます。

**画像データを表示するための関数 plt.imshow()**

plt.imshow() は2次元配列を画像として表示する関数です。この関数は、科学的ビジュアライゼーションで頻繁に使用されます。例えば、行列の可視化や画像処理結果の表示に使われます。

```python
import matplotlib.pyplot as plt
import numpy as np

# 10x10のランダムな行列を生成
data = np.random.rand(10, 10)

# 行列を画像として表示
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()  # カラーバーを追加
plt.title("Image Plot of Random Data")
plt.show()
```

このコードでは、ランダムな値を含む10x10の行列を生成し、imshowを使って画像として表示しています。cmap引数でカラーマップを指定し、interpolation引数でピクセル間の補間方法を指定しています。

**その他の便利な関数**

**plt.colorbar()**

この関数は、imshow()で表示した画像に対してカラーバー（色のスケール）を追加します。データの値に対応する色を示すことで、画像の解釈を助けます。

**plt.hist2d()**

2次元のヒストグラムを作成します。これは、二つのデータセットの共同分布をビジュアライズするのに使います。

```python
x = np.random.randn(10000)
y = np.random.randn(10000)

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar(label='counts in bin')
plt.title("2D Histogram")
plt.show()
```

**図のサイズやグラフの数を指定する**

matplotlibで複数のグラフを1つの図（Figure）内に配置するには、plt.figure() 関数とサブプロットを用います。plt.figure() は新しい図を作成し、サブプロット（plt.subplot()）を使ってその図の中に複数のグラフを配置できます。

**plt.figure()**

plt.figure() 関数は新しい図を作成するために使います。この関数はオプションで図のサイズや解像度を指定できます。

**サブプロットの追加 (plt.subplot())**

plt.subplot() 関数は、図内に複数のグラフ（サブプロット）を配置するために使用します。この関数には3つの主要な引数があります:
*   nrows: 図に含まれる行の数
*   ncols: 図に含まれる列の数
*   index: 配置するサブプロットの位置（左上から右下へのインデックス）

**実際のコード例**

以下の例では、1つの図内に4つの異なるグラフ（2行2列）を配置しています。

```python
import matplotlib.pyplot as plt
import numpy as np

# 新しい図を作成し、そのサイズを指定
plt.figure(figsize=(10, 8))

# 1つ目のサブプロット（行1, 列1の位置）
plt.subplot(2, 2, 1)  # 2行2列の1番目
plt.plot(np.random.rand(10))
plt.title("Line Plot")

# 2つ目のサブプロット（行1, 列2の位置）
plt.subplot(2, 2, 2)  # 2行2列の2番目
plt.scatter(np.arange(10), np.random.rand(10))
plt.title("Scatter Plot")

# 3つ目のサブプロット（行2, 列1の位置）
plt.subplot(2, 2, 3)  # 2行2列の3番目
plt.bar(['A', 'B', 'C'], [5, 7, 3])
plt.title("Bar Chart")

# 4つ目のサブプロット（行2, 列2の位置）
plt.subplot(2, 2, 4)  # 2行2列の4番目
plt.hist(np.random.randn(1000), bins=20)
plt.title("Histogram")

# 図全体のタイトルと余白の調整
plt.suptitle("Multiple Plots Example")
plt.tight_layout(pad=3.0)  # サブプロット間のパディングを調整
plt.show()
```

このコードでは、plt.figure() で新しい図を作成し、plt.subplot() を使ってその図内に4つの異なるグラフを配置しています。plt.suptitle() は全体の図にタイトルを付けるために使用し、plt.tight_layout() はサブプロット間の重なりを防ぎつつ適切な間隔を保つために役立ちます。

**応用**

**より細かい設定をするfig, axの使い方**

これまでの plt 関数を使う方法と比べて、fig, ax の使い方はより明示的で、より細かなコントロールが可能になります。

まず、plt.subplots() 関数を使って fig と ax を作成します。この関数は Figure と Axes オブジェクトを返します。ここで、Axes オブジェクトはグラフの実際のプロットエリアを表し、Figure はこのプロットエリアを含む全体のウィンドウまたはページを表します。

```python
import matplotlib.pyplot as plt
import numpy as np

# FigureとAxesを作成
fig, ax = plt.subplots()

# Axesに対してグラフをプロット
ax.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])

# タイトル、軸ラベルを設定
ax.set_title("Simple Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")

# 表示
plt.show()
```

**複数のサブプロットの例**

plt.subplots() に行数と列数を指定することで、複数のグラフを含む図を作成できます。

```python
fig, axs = plt.subplots(2, 2)  # 2行2列のサブプロット

# 各サブプロットにデータをプロット
axs[0, 0].plot(np.random.rand(10))
axs[0, 0].set_title('Subplot 1')

axs[0, 1].plot(np.random.rand(10), 'tab:orange')
axs[0, 1].set_title('Subplot 2')

axs[1, 0].plot(np.random.rand(10), 'tab:green')
axs[1, 0].set_title('Subplot 3')

axs[1, 1].plot(np.random.rand(10), 'tab:red')
axs[1, 1].set_title('Subplot 4')

# タイトルと軸ラベルの設定、表示
fig.suptitle('Multiple Subplots')
plt.tight_layout()
plt.show()
```

**fig, ax アプローチと plt アプローチの違い**



1.   コントロールの精度: fig, ax アプローチでは、複数の ax インスタンスを個別に扱うことができるため、グラフごとに細かく設定を変更することができます。
2.   再利用性: fig, ax アプローチでは、Axes オブジェクトを関数に渡したり、後で再利用することが容易です。これにより、コードのモジュール性と再利用性が向上します。
3.  複雑なレイアウト: 複雑なグラフのレイアウトを容易に構築できます。たとえば、グリッドシステムの中に異なるサイズや形状のサブプロットを配置することが可能です。
